````python
# Tu solucion
import pandas as pd
df = pd.read_csv('yelp.csv')
df
````

2. Ignoramos el archivo `yelp.csv` y construye un DataFrame de `yelp.json`. Esto implica leer los datos en Python, decodificar el JSON, convertirlo en un DataFrame y agregar columnas individuales para cada uno de los tipos de votos.

````python
# Tu solucion
import pandas as pd
import json
df = pd.DataFrame(data=[],columns=['business_id', 'date',  'review_id', 'stars', 'text','type', 'user_id','cool', 'useful','funny'])
for line in open('yelp.json', 'r'):
    df = df.append({'business_id': json.loads(line)['business_id'],'date': json.loads(line)['date'],'review_id': json.loads(line)['review_id'],'stars': json.loads(line)['stars'],'text': json.loads(line)['text'],'type': json.loads(line)['type'],'user_id': json.loads(line)['user_id'],'cool': json.loads(line)['votes']['cool'],'useful': json.loads(line)['votes']['useful'],'funny': json.loads(line)['votes']['funny']}, ignore_index=True,sort=None)
df
````

3. Explora la relación entre cada uno de los tipos de votos (cool/useful/funny) y la cantidad de estrellas.

````python
# Tu solucion
df[['cool','useful','funny', 'stars']]
````

4. Define `cool/ useful/funny` como características y `stars` como respuesta.

````python
# Tu solucion
X= df[['cool', 'useful','funny']]
y= df[['stars']]
````

5. Ajustamos un modelo de regresión lineal e interpreta los coeficientes. ¿Los coeficientes tienen sentido intuitivo para ti?. Explora el sitio web de Yelp para ver si se puede detectar tendencias similares.
    

````python
# Tu solucion
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics

linreg = LinearRegression()
linreg.fit(X, y)

print( linreg.coef_)

````

6. Evalua el modelo dividiendo los datos en  conjuntos de entrenamiento y prueba y calculando el RMSE. ¿El RMSE tiene un sentido intuitivo para ti?
    

````python
# Tu solucion
import numpy as np

X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, random_state=1)
linreg.fit(X_entrenamiento, y_entrenamiento)
y_predic = linreg.predict(X_prueba)
#RMSE
np.sqrt(metrics.mean_squared_error(y_prueba, y_predic))
````

7. Intenta eliminar algunas de las características y analizar  si el RMSE mejora.
   

````python
# Tu solucion
X= df[['useful','funny']]
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, random_state=1)
linreg.fit(X_entrenamiento, y_entrenamiento)
y_predic = linreg.predict(X_prueba)
#RMSE
np.sqrt(metrics.mean_squared_error(y_prueba, y_predic))
````

- Divide los datos en el conjunto de entrenamiento y prueba usando `train_test_split`.
- Entrena y ajusta un modelo de regresión logística en el conjunto de entrenamiento.
- Pronostica valores para los datos de prueba.
- Crea un informe de clasificación para el modelo. Utiliza `classification_report`.

````python
# Tu solucion
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
logisticRegr = LogisticRegression()
datos.drop(['Ad Topic Line', 'City', 'Country', 'Timestamp'], axis=1, inplace=True)
X = datos.drop(['Clicked on Ad'], axis = 1)
y = datos['Clicked on Ad']
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, random_state=1)
logisticRegr.fit(X_entrenamiento, y_entrenamiento)
y_pred=logisticRegr.predict(X_prueba)
print(classification_report(y_prueba, y_pred))

````

 Naive bayes es un algoritmo eager ya que no necesita volver a usar los datos de aprendizaje para predecir un nuevo valor a diferencia de k-vecinos que es un algoritmo lazy porque para una nueva instancia necesita reevaluar todos los datos de entrenamiento. Pero a pesar de ser eager,construir el modelo de naive bayes es mas costoso computacionalmente que construir un modelo lazy como k-vecinos.
